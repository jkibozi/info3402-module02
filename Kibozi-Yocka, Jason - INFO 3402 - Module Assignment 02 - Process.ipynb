## Dataset #1
[Motor Vehicle Occupant Death Rate, by Age and Gender, 2012 & 2014, All States](https://catalog.data.gov/dataset/motor-vehicle-occupant-death-rate-by-age-and-gender-2012-2014-all-states)

```python
# read in dataset as dataframe
mvoDF = pd.read_csv("MVO_DRate_12-14.csv")

mvoDF.head()
```

```python
# drop empty first row and 'State' column in dataframe
mvoDF.drop(index=0,inplace=True)
mvoDF.drop(columns=['State'],inplace=True)
mvoDF.reset_index(drop=True,inplace=True)

mvoDF.head()
```

```python
# split 'Location' column from dataframe into 'State', 'Latitude', and 'Longitude' columns
locList = mvoDF['Location']
locDF = pd.DataFrame(columns=['State','Latitude','Longitude'])
count = 0

for location in locList:
    location = location.replace('\n','').replace(', ',',').replace('(',',').replace(')','')
    location = location.split(',')
    state = location[0]
    latitude = location[1]
    longitude = location[2]
    locDF.loc[count] = [state,latitude,longitude]
    count += 1

locDF.head()
```

```python
# add 'State', 'Latitude', and 'Longitude' columns to dataframe
mvoDF.drop(columns=['Location'],inplace=True)
mvoDF.insert(0, 'Longitude',locDF['Longitude'],True)
mvoDF.insert(0, 'Latitude',locDF['Latitude'],True)
mvoDF.insert(0, 'State',locDF['State'],True)

mvoDF.head()
```

```python
# drop rows in dataframe where 'State' = 'United States' or 'District of Columbia'
mvoDF = mvoDF[mvoDF['State'] != 'United States']
mvoDF = mvoDF[mvoDF['State'] != 'District of Columbia']
mvoDF.reset_index(drop=True,inplace=True)

mvoDF['State'].unique()
```

#### Dataset #3
[Total Census: All Geocodes 2016](https://github.com/GL-Li/totalcensus/blob/master/data_raw/all-geocodes-v2016%20.xlsx)

```python
# read in dataset as dataframe
geoDF = pd.read_excel('all-geocodes-v2016.xlsx')

# rename columns
geoDF.columns = ['Summary Level','State Code','County Code','County Subdivision','Place Code','Consolidtated City Code',
                 'Area Name']

geoDF.head()
```

```python
# convert state codes to state names

# Part 1: Create a dictionary with every state and its corresponding state code
stateDic = {'Alaska':'','Alabama':'','Arkansas':'','Arizona':'','California':'',
            'Colorado':'','Connecticut':'','Delaware':'','Florida':'','Georgia':'',
            'Hawaii':'','Iowa':'','Idaho':'','Illinois':'','Indiana':'',
            'Kansas':'','Kentucky':'','Louisiana':'','Massachusetts':'','Maryland':'',
            'Maine':'','Michigan':'','Minnesota':'','Missouri':'','Mississippi':'',
            'Montana':'','North Carolina':'','North Dakota':'','Nebraska':'','New Hampshire':'',
            'New Jersey':'','New Mexico':'','Nevada':'','New York':'','Ohio':'',
            'Oklahoma':'','Oregon':'','Pennsylvania':'','Rhode Island':'','South Carolina':'',
            'South Dakota':'','Tennessee':'','Texas':'','Utah':'','Virginia':'',
            'Vermont':'','Washington':'','Wisconsin':'','West Virginia':'','Wyoming':''
           }
# https://code.activestate.com/recipes/577305-python-dictionary-of-us-states-and-territories/

for state in stateDic:
    name = geoDF[geoDF['Area Name'] == state]
    stateDic[state] = name['State Code'].item()

stateDic = dict((v,k) for k,v in stateDic.items())
# https://stackoverflow.com/questions/1031851/python-best-way-to-exchange-keys-with-values-in-a-dictionary

print(stateDic)
```

```python
# convert state codes to state names

# Part 2: map state names from dictionary to state codes in dataframe 'STATE' column
v12DF['STATE'] = v12DF['STATE'].map(stateDic)
# https://stackoverflow.com/questions/20250771/remap-values-in-pandas-column-with-a-dict

v12DF.head()
```

```python
# check for NaN values causesed by .map()
v12DF['STATE'].unique()

# looks like there are NaN values (look between Deleware and Florida below)
```

```python
# drop rows in dataframe if 'STATE' column contains NaN
v12DF.dropna(axis=0,subset=['STATE'],inplace=True)
v12DF.reset_index(drop=True,inplace=True)

v12DF['STATE'].unique()
# looks like that got rid of them
```

```python
# convert county codes to county names

# Part 1: create dictionary of county codes and names
countyDF = geoDF[geoDF['County Code'] != 0]
countyDF = countyDF[countyDF['County Subdivision'] == 0]

countyDF.reset_index(drop=True,inplace=True)

countyDF.head()
```

```python
countyDict = {}

for i, row in countyDF.iterrows():
    countyDict[row['County Code']] = row['Area Name']
```

```python
# # convert county codes to county names

# Part 2: map county names from dictionary to county codes in dataframe 'COUNTY' column
v12DF['COUNTY'] = v12DF['COUNTY'].map(countyDict)
# https://stackoverflow.com/questions/20250771/remap-values-in-pandas-column-with-a-dict

v12DF.head()
```

```python
# check for NaN values causesed by .map()
v12DF['COUNTY'].unique()

# looks like there are NaN values (look between La Paz County and Toa Alta Municipio below)
```

```python
# drop rows in dataframe if 'STATE' column contains NaN
v12DF.dropna(axis=0,subset=['COUNTY'],inplace=True)
v12DF.reset_index(drop=True,inplace=True)

v12DF['COUNTY'].unique()
# looks like that got rid of them
```

```python
# replace month number with month name
MonthDict = {1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',
             11:'November',12:'December'}

v12DF['MONTH'] = v12DF['MONTH'].map(MonthDict)

v12DF['MONTH'].unique()
```

```python
v12DF.head()
```

```python
# combine 'DAY' and 'MONTH' into 'DATE' column
dateList = []

for i, row in v12DF.iterrows():
    date = str(row['MONTH'])+' '+str(row['DAY'])+', '+'2012'
    dateList.append(date)

v12DF.drop(columns=['DAY','MONTH'],inplace=True)
v12DF.insert(4, 'DATE',dateList,True)
v12DF.head()
```

```python
# combine 'HOUR' and 'MINUTE' into 'TIME' column
timeList = []

for i, row in v12DF.iterrows():
    time = str(row['HOUR'])+':'+str(row['MINUTE'])
    timeList.append(time)

v12DF.drop(columns=['HOUR','MINUTE'],inplace=True)
v12DF.insert(5, 'TIME',timeList,True)
v12DF.head()
```
